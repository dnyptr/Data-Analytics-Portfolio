## Load Data

```python
import pandas as pd
df = pd.read_csv('online_retail_II.csv')
df
```

## Section 1

### Create New Feature: Year

```python
df['Year'] = pd.DatetimeIndex(df['InvoiceDate']).year
df
```

### Filtering Data

```python
df[ (df['Quantity'] > 1) & (df['Invoice'] != '%C%') ]
```

### Create New Feature: Revenue

```python
df['Revenue'] = df['Quantity'] * df['Price']
df
```

### Average of Revenue per Year

```python
rpy = df.groupby('Year')['Revenue'].mean().reset_index()
rpy
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.DataFrame(rpy, columns = ["Year","Revenue"])

data.plot(x="Year",y="Revenue", kind="line", figsize = (8,8))
```

### Interpretation

# The revenue in 2009 increases to 18.15 in 2010, and decreased to 18.01 in 2011. Additionally, the revenue in 2010 is the highest after 2011’s revenue, with 2009 having the lowest of them.
